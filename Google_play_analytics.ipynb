```python
def size_processor(i):
    if str(i)[-1] == 'M':
        s = float(str(i).replace('M', ' '))
        s = s * 1024
        return s
    elif str(i)[-1] == 'K':
        s = float(str(i).replace('K', ' '))
        return s
    else:
        return str(np.nan)
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_processor)
```

## Make a generalized code to remove characters from coloumns 

```python
def col_remover():
    columns = ["Price","Installs"]
    char_remover = ["+",",","$"]
    for col in columns:
        df_copy[col] = df_copy[col].astype(str)  # it is necessary to convert a column datatype to str first
    for char in char_remover:
        for col in columns:
            df_copy[col] = df_copy[col].str.replace(char, "", regex=False)
    return df_copy[columns]
```

```python
col_remover()   ## Whenever define a function make sure to call it back
```
