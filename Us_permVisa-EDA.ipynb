#Computer Systems Analyst - V , Computer Systems Analyst - IV and so on, have different job title

#software engineer , SOFTWARE ENGINEER, Sr. Software Engineer >> all of this should come in same profile  

```python
df.job_info_job_title = df.job_info_job_title.str.lower()
```

```python
df['job_info_job_title'].value_counts(dropna=True)[:10]
```

```python
split = ['-', '/', 'ii']

df['job_info_job_title'] = df['job_info_job_title'].str.split('-').str[0]
df['job_info_job_title'] = df['job_info_job_title'].str.split('/').str[0]
df['job_info_job_title'] = df['job_info_job_title'].str.split('ii').str[0]
df['job_info_job_title'] = df['job_info_job_title'].str.strip()
df['job_info_job_title'] = df['job_info_job_title'].str.replace('sr.', 'senior')
df['job_info_job_title'] = df['job_info_job_title'].replace(['NaN','NaT', 'nan'], np.nan)



```

```python
top_10_jobs = df['job_info_job_title'].value_counts(dropna=True)[:10]
```

```python
colors = plt.cm.viridis(range(10))
top_10_jobs.plot(kind = 'bar', color = colors, x = 'job title', y = 'No. of visa Applicants')
```
