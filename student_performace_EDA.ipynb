Key Insights


This table shows the frequency distribution of students across combinations of Workday (Dalc) and Weekend (Walc) alcohol consumption, both rated 1 (very low) to 5 (very high).


Diagonal trend — Most students fall along or above the diagonal (Walc ≥ Dalc), meaning students tend to drink more on weekends than on weekdays, which is expected behavior.
Majority are low consumers — 241 students have Dalc=1 & Walc=1, meaning the largest group drinks very little on both days.
Strong correlation — Students with high Dalc (4 or 5) also tend to have high Walc (e.g., Dalc=5, Walc=5 has 15 students), showing heavy weekday drinkers are also heavy weekend drinkers.
Very few heavy weekday drinkers — Dalc of 3, 4, or 5 has very few students, but those who do drink heavily on weekdays almost always drink heavily on weekends too.
Low Dalc, moderate Walc — Many students with Dalc=1 still drink moderately on weekends (113 at Walc=2, 64 at Walc=3), showing weekend-only moderate drinking is common.

```python
def detect_outliers(columns):
    outlier_indices = []
    for col in columns:
        ## Find Q1,Q2 and IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3-Q1
        ## Mask is a condition that returns True False values for a particular coloumn
        mask = (df[col] >= (Q1-1.5*IQR)) & (df[col] <= (Q3+1.5*IQR))
## np.argwhere is a NumPy function that finds the indices of all non-zero elements in an array. 
        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)

    return np.unique(np.concatenate(outlier_indices).ravel())

```
