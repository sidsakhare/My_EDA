Key Observations
Class 0 ( "No Ckd" / Healthy)

```python
sns.violinplot(x = 'class', y = 'blood_urea', data = df , palette= 'pastel')
```

Key Observations
Class 0 (Healthy / No CKD)

Values tightly clustered between ~10–60
Violin is short and wide → low variance, concentrated distribution
Median is low (~25–30), very little spread
Almost no extreme values

Class 1 (Diseased / CKD)

Values range from near 0 up to ~430
Violin is tall and narrow at top → heavily right-skewed with a long upper tail
Median (~60–70) is notably higher than Class 0
The wide middle section indicates high density around 50–100, but extreme cases exist

```python
diabetes_hypertension = pd.crosstab(df.hypertension , df.diabetes_mellitus)
```

```python
diabetes_hypertension.plot(kind = 'bar', stacked = False)
```

Key Observations
Among patients WITHOUT hypertension (left group):

~220 do not have diabetes
~30 do have diabetes
Diabetes is relatively rare (~12% prevalence)

Among patients WITH hypertension (right group):

~42 do not have diabetes
~107 do have diabetes
Diabetes is quite common (~72% prevalence)


Strong association between hypertension and diabetes — patients with hypertension are far more likely to also have diabetes mellitus.

```python
cols = ['age', 'blood_pressure', 'blood_glucose_random', 'serum_creatinine', 'class']

g = sns.PairGrid(df[cols], hue='class', palette = 'coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap = 'Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
plt.title("PairGrid for selected columns")
plt.show()
```

```python
plt.figure(figsize= [20,15])
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot= True, cmap = 'coolwarm')
plt.show()
```

```python
sns.swarmplot(x = 'age', y = 'hypertension',data = df, hue = 'class', palette= 'pastel')
```

#### difference between scatterplot and swarmplot is datapoints in swarm plot are  not overlapped at all while datapoints in scatterplot may be overlapped 

```python
fig = px.scatter(df, y = 'age', x = 'hypertension', color = 'class', hover_data= ['age', 'hypertension'], title = "interactive scatterplot with hover information")
fig.show()
```

```python
fig = px.scatter_3d(df, y = 'age', x = 'hypertension',z = 'blood_glucose_random',  color = 'class', hover_data= ['age', 'hypertension','blood_glucose_random'], title = "interactive scatterplot with hover information")
fig.show()
```
